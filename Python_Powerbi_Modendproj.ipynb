```python
df.rename(columns = {'Country/Region':'Country'}, inplace=True)
```

```python


Top_Countries=df.groupby('Country').agg({'Confirmed':'sum'}).sort_values(by="Confirmed",ascending=False)
Top_Countries
```

```python


# To plot the top affected countries by range 10
Top_confirmed=Top_Countries.iloc[:10]
Top_confirmed

```

```python
labels = Top_Countries.index[:10]  # Top 10 countries
sizes = Top_Countries['Confirmed'][:10]  # Corresponding confirmed case numbers

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries by Confirmed Cases')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

```python
# to view the count of recoveries
top_Recovery=df.groupby('Country').agg({'Recovered':'sum'}).sort_values(by="Recovered",ascending=False)
top_Recovery
```

```python
top10_recovery=top_Recovery.iloc[:10]
top10_recovery
```

```python
#  ploting the countries with highest recovery


top10_recovery.plot(kind='bar')
plt.xlabel( "Recovery Rate" )

# Set label for y-axis
plt.ylabel( "Top Countries" )

# Set title for figure
plt.title( "Top 10 countries with highest recovery rate" )
```

```python


Country_confm_recovery= df.groupby('Country').agg({'Confirmed': 'sum', 'Recovered': 'sum', 'Deaths': 'sum'}).reset_index()
                                  
Country_confm_recovery
df1= pd.DataFrame(Country_confm_recovery)
df1
```

```python
top10_death=df.groupby('Country').agg({'Deaths':'sum'}).sort_values(by="Deaths",ascending=False)
top10_death
```

```python
Top_death=top10_death.iloc[:10]
Top_death
```

```python

#  ploting the top models
plt.figure(figsize=(8,6))
plt.xticks(rotation=75)
sns.barplot(x=Top_death.index, y='Deaths',data=Top_death)
plt.xlabel( "Countries")

# Set label for y-axis
plt.ylabel( "Count of Deaths" )

# Set title for figure
plt.title( "Highest Death reported countries " )
plt.show()


 
```
